## Part 2: Creating your (Parent) Class 

#### 2.1: Creating the Class (9 Points)

&#9989; Create a class called `Production`. The Class should have the attributes `title`, `release_year`, and `genre`. The class should also have `get` and `set` methods for each of the three attributes.

```python

class Production:
    
    def __init__(self , title , year , genre):
        """ Initialize the production object.
        Inputs are
        title : the title of the show/movie (string)
        release_year : the release_year of the show/movie (integer)
        genre : the genre of the show/movie (string)
        """
        self.title = title
        self.release_year = year
        self.genre = genre
        
        
    def set_title(self,title):
        self.title = title
        
    def set_year(self,year):
        self.release_year = year
        
    def set_genre(self,genre):
        self.genre = genre
        
    def get_title(self):
        return self.title
    
    def get_year(self):
        return self.release_year
    
    def get_genre(self):
        return self.genre
    
```

Make sure that each of the child classes has an `__init__` function that takes all of the necessary information to create an instance of that class. 

**Note:** Your child classes should use the `super()` function.

```python
#Write your (child) classes here
class Movie(Production):
    def _init_(self, title , year , genre):
        super()._init_(title , year , genre)
    def set_director(self,director):
        self.director = director
    def get_director(self):
        return self.director
class TV_Show(Production):
    def _init_(self, title , year , genre):
        super()._init_(title , year , genre)
    def set_seasons(self,seasons):
        self.seasons = seasons
    def get_seasons(self):
        return self.seasons
    
```

&#9989;  Take the code that you wrote in the previous problem and turn it into a method for the `Production` class. Run this method within the `__init__` method so that every `Production` object comes into being with the genres split apart into a list.

```python
#Write your code here
class Production:
    
    def __init__(self , title , year , genre):
        """ Initialize the production object.
        Inputs are
        title : the title of the show/movie (string)
        release_year : the release_year of the show/movie (integer)
        genre : the genre of the show/movie (string)
        """
        self.title = title
        self.release_year = year
        self.genre = []
        for i in genre:
            i = i.split(',')
            self.genre.append(i)

            
            
            
            
            
        
        
    def set_title(self,title):
        self.title = title
        
    def set_year(self,year):
        self.release_year = year
        
    def set_genre(self,genre):
        self.genre = genre
        
    def get_title(self):
        return self.title
    
    def get_year(self):
        return self.release_year
    
    def get_genre(self):
        return self.genre
```
